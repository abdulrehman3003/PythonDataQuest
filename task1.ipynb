```python
from matplotlib import pyplot as plt
import pandas as pd
import os
```

```python
task1_file=r"Data Quest Task1.csv"
task1 = pd.read_csv(task1_file,encoding="latin-1")
task1
```

```python
task1['dteday'] = pd.to_datetime(task1['dteday']) 
```

```python


# Time-Series Plot
if 'dteday' in task1.columns and 'hum' in task1.columns:  
        plt.figure(figsize=(10, 5))
        plt.plot(task1['dteday'], task1['atemp'], label='Average Temperature over time') 
        plt.plot(task1['dteday'], task1['hum'], label='Average Humidity over time') 
        plt.plot(task1['dteday'], task1['windspeed'], label='Average windspeed over time') 
        plt.title('Time-Series of windspeed, average temperature, humidity')
        plt.xlabel('Date')
        plt.ylabel('Time-Series of windspeed, average temperature, humidity')
        plt.legend()
        plt.show()

 # Histogram
plt.figure(figsize=(8, 5))
plt.hist(task1['atemp'], bins=20, color='skyblue', edgecolor='black')   
plt.title('Distribution of Average Temperature')
plt.xlabel('Days')
plt.ylabel('Average Temperature')
plt.show()

# Scatter Plot

plt.figure(figsize=(8, 5))
plt.scatter(task1['registered'], task1['cnt'])
plt.title('Scatter Plot')
plt.xlabel('registered')
plt.ylabel('registerd')
plt.show()
```
