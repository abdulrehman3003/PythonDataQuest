```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
task2=pd.read_csv(r"C:\Users\abdul\Documents\PythonDataQuest\Data Quest Task2.csv", encoding="latin-1")
task2.head()
```

```python
task2["Gender"]=task2["Gender"].str.replace(")","")
task2["Gender"]=task2["Gender"].str.replace("(","")
task2['Nationality']=task2['Nationality'].str.replace(")",(""))
task2['Nationality']=task2['Nationality'].str.replace("(","")
task2["BeginDate"]=task2["BeginDate"].replace("-","")
task2.head()
```

```python
task2["BeginDate"] = pd.to_numeric(task2["BeginDate"]).astype("Int64")
task2["EndDate"] = pd.to_numeric(task2["EndDate"]).astype("Int64")

task2["actual_age"] = task2["EndDate"] - task2["BeginDate"]
task2["actual_age"] = task2["actual_age"].astype(str).replace(r"-","")  
task2["Date"] = task2["Date"].astype(str).str.replace(r"-", "", regex=True)
task2.head()

```
